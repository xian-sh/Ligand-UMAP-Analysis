# Molecular Feature Case Demo

This notebook runs a small end-to-end demo with RDKit fingerprints, PCA, UMAP, and KMeans. It creates a tiny SMILES table locally, so it can be used without the original notebook data file.

```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
repo_root = cwd.parent if cwd.name == "scripts" else cwd
sys.path.insert(0, str(repo_root))

repo_root
```

```python
import pandas as pd

demo_data = pd.DataFrame(
    {
        "ID": [f"mol_{i:02d}" for i in range(1, 13)],
        "SMILES": [
            "CCO",
            "CCCO",
            "CCCCO",
            "CC(=O)O",
            "CCC(=O)O",
            "c1ccccc1",
            "Cc1ccccc1",
            "Oc1ccccc1",
            "CCN(CC)CC",
            "CCOC(=O)C",
            "C1CCCCC1",
            "CCS",
        ],
    }
)

demo_csv = repo_root / "scripts" / "demo_molecules.csv"
demo_data.to_csv(demo_csv, index=False)
demo_data
```

Run the pipeline with fingerprint features. For ChemBERT or RDKit descriptors, change `feature_type` to `"chembert"` or `"descriptor"`.

```python
from molecular_feature_case import Config, run_pipeline

cfg = Config(
    input=demo_csv,
    output_dir=repo_root / "scripts" / "demo_outputs",
    feature_type="fingerprint",
    fp_bits=256,
    pca_components=8,
    umap_components=2,
    umap_neighbors=3,
    min_clusters=2,
    max_clusters=4,
    force_clusters=3,
    ids_to_mark=("mol_01", "mol_06"),
)

run_pipeline(cfg)
```

```python
result_path = cfg.output_dir / "clustered_molecules_fingerprint.csv"
result = pd.read_csv(result_path)
result.head()
```
